# Section 17 — Common errors and how to fix them

When you are learning Python, errors are normal. In this notebook, most practice cells are wrapped so that common errors show a short hint instead of a long traceback.

Common examples:

| Error type | Meaning | Common fix |
|---|---|---|
| `NameError` | Python does not know a word you typed | Put text in quotes, or run the cell where the variable was created |
| `KeyError` | A column name is wrong | Check spelling and capital letters |
| `TypeError` | Wrong data type | Check if something should be text, number, list, or DataFrame |
| `SyntaxError` | Python grammar is wrong | Check brackets, quotes, commas, and spelling |

Important: `SyntaxError` can still show a red message because Python cannot understand the cell at all. If that happens, check quotes, brackets, and commas first.
