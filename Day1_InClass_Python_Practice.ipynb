## Section 2 — Lists

A **list** holds multiple values in order. You create one with square brackets `[...]`.

```python
# A list of fruits
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(fruits)
```

You access items by their **position**. Python is **0-indexed**, which means counting starts at 0, not 1. The first item is at position 0.

```python
print("First fruit:", fruits[0])
print("Second fruit:", fruits[1])
print("Last fruit:", fruits[-1])   # negative numbers count from the end

print("How many fruits?", len(fruits))
```
