## A peek at what's coming — EDA 🔍

So far we've **loaded** the data, **described** it, and **cleaned** it. That's the setup. **Exploratory Data Analysis** (EDA) is what comes next — we start asking questions and let the data answer.

Here's one of the first questions a scientist would ask: **do the three species differ in size?**

We can answer it in one line: group by species, then take the mean of each numeric column.

```python
df.groupby("species")[["bill_length_mm", "flipper_length_mm", "body_mass_g"]].mean().round(1)
```

The **body mass** column shows the three species are clearly different sizes — **Gentoo** penguins are the heaviest by far.

A table only tells part of the story, though. Let's **see** the data. We'll start simple: a scatter plot of just **one** species — the Adelie penguins.

```python
# Scatter plot of just the Adelie penguins
import matplotlib.pyplot as plt

adelie = df[df["species"] == "Adelie"]

plt.figure(figsize=(8, 5))
plt.scatter(adelie["flipper_length_mm"], adelie["body_mass_g"], alpha=0.7)
plt.xlabel("Flipper length (mm)")
plt.ylabel("Body mass (g)")
plt.title("Adelie penguins — flipper length vs body mass")
plt.show()
```

Each dot is one Adelie penguin. There's a clear trend — longer flippers tend to go with heavier birds. Good.

Now the interesting question: **what if we add the other two species to the same plot?**

```python
# Same plot, but now showing all three species — one color per species
plt.figure(figsize=(8, 5))
for species in df["species"].unique():
    mask = df["species"] == species
    plt.scatter(df.loc[mask, "flipper_length_mm"],
                df.loc[mask, "body_mass_g"],
                label=species, alpha=0.7)

plt.xlabel("Flipper length (mm)")
plt.ylabel("Body mass (g)")
plt.title("All three species — the clusters separate")
plt.legend()
plt.show()
```

See those three clusters? **Gentoo** sit up and to the right (long flippers, heavy). **Adelie** and **Chinstrap** overlap more on this view, but they're distinguishable by other measurements (bill size especially — we'll see that in Day 3).

This is the heart of EDA: looking at data until patterns reveal themselves. And those patterns suggest something powerful:

> If the species look this different from each other, **maybe a model could learn to tell them apart from the measurements alone.**
